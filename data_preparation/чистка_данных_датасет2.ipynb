```python

df = pd.read_csv("новые_протесты_рус3.csv")

def print_unique_values(column_name):

    clean_values = df[column_name].dropna().astype(str)
    clean_values = clean_values[clean_values.str.strip() != '']
    
    unique_values = sorted(clean_values.unique())
    
    print(f"Уникальные значения {column_name} ({len(unique_values)}):")
    for idx, value in enumerate(unique_values, 1):
        print(f"{idx}. {value}")

print_unique_values('region_name')
print("\n" + "="*50 + "\n")
print_unique_values('municipality')
```

```python

import re


df = pd.read_csv('новые_протесты_рус3.csv')

# Эталонный список
correct_regions = [
    'Тверская', 'Магаданская', 'Смоленская', 'Чукотский', 'Тыва', 'Татарстан',
    'Ставропольский', 'Астраханская', 'Новгородская', 'Коми', 'Костромская',
    'Нижегородская', 'Забайкальский', 'Саха', 'Адыгея', 'Хакасия',
    'Карачаево-Черкесская', 'Вологодская', 'Тамбовская', 'Красноярский',
    'Тюменская', 'Самарская', 'Пензенская', 'Чувашская', 'Волгоградская',
    'Оренбургская', 'Башкортостан', 'Краснодарский', 'Новосибирская',
    'Воронежская', 'Крым', 'Дагестан', 'Хабаровский', 'Тульская', 'Карелия',
    'Московская', 'Ростовская', 'Чеченская', 'Калужская', 'Ленинградская',
    'Ханты-Мансийский', 'Челябинская', 'Саратовская', 'Рязанская', 'Удмуртская',
    'Липецкая', 'Бурятия', 'Мордовия', 'Кировская', 'Курганская', 'Омская',
    'Марий Эл', 'Курская', 'Иркутская', 'Ямало-Ненецкий', 'Белгородская',
    'Алтайский', 'Кемеровская', 'Ингушетия', 'Амурская', 'Алтай', 'Брянская',
    'Северная Осетия-Алания', 'Мурманская', 'Свердловская', 'Ульяновская',
    'Владимирская', 'Сахалинская', 'Кабардино-Балкарская', 'Пермский',
    'Томская', 'Приморский', 'Псковская', 'Архангельская', 'Камчатский',
    'Орловская', 'Ненецкий', 'Еврейская', 'Ярославская', 'Ивановская',
    'Калмыкия', 'Калининградская', 'Севастополь'
]

# Строим словарь префиксов
prefix_map = {region[:4].lower(): region for region in correct_regions}

# Функция для замены по первым четырём буквам
def normalize_region(name: str) -> str:
    """
    Если первые четыре буквы (без учёта регистра) совпадают с ключом,
    возвращает эталонное имя, иначе — оригинал.
    """
    if not isinstance(name, str) or len(name) < 4:
        return name
    prefix = name[:4].lower()
    return prefix_map.get(prefix, name)


df['region_name'] = df['region_name'].apply(normalize_region)


df.to_csv('новые_протесты_рус3.csv', index=False)
```
